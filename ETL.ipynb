vamos a transformar los valores de la columna HH a tipo numérico, ya que estan tipo object.

```python
# Cambio el dato horas  a tipo de dato numerico y luego a entero 
homicidios_hecho["HH"] = pd.to_numeric(homicidios_hecho ["HH"], errors="coerce")
homicidios_hecho["HH"] = homicidios_hecho["HH"].astype("Int64")
```

Las coordenadas X y Y debemos  transformar el tipo de dato ya que  estan en tipo Objeto.  estan deben para a tipo float(), tambien debemos revisar que  posicionen no son válidas.

```python
# Filtrar filas con posiciones no válidas
valid_positions = ~(homicidios_hecho['pos x'].str.contains('[^0-9.,-]') | homicidios_hecho['pos y'].str.contains('[^0-9.,-]') )

```

```python
# Establecer posiciones no válidas en NaN
homicidios_hecho.loc[~valid_positions, ['pos x', 'pos y']] = np.nan
```

```python
homicidios_hecho['pos y'] = pd.to_numeric(homicidios_hecho['pos y'].str.replace(',', '.'), errors='coerce')
```

Aca transformamos los valores de latitud y  longitud a tipo float.

```python
homicidios_hecho['pos x'] = pd.to_numeric(homicidios_hecho['pos x'].astype(str).str.replace(',', '.'), errors='coerce').astype(float)
homicidios_hecho['pos y'] = pd.to_numeric(homicidios_hecho['pos y'].astype(str).str.replace(',', '.'), errors='coerce').astype(float)
```

```python
homicidios_hecho.info()
```

Creamos una columna llamada semestre, esto con el fin de trabajar los KPI.

```python

homicidios_hecho['semestre'] = [1 if month <= 6 else 2 for month in homicidios_hecho['FECHA'].dt.month]
```

Debemos renombrar el nombre de la columna COMUNA a COMUNAS  para poder realizar el merge.

Vamos a crear una nueva columna con el nombre del  dia de la semana que corresponde el siniestro, con esto podremos  ver como  es el comportamiento de  los siniestros por cada dia, podriamos estudiar los fines de semana en conjunto tambien y comparar con los dias de semana. 

```python
#Creo una columna nueva `Día semana` que contenga el nombre del día de la semana
homicidios_hecho['DIA SEMANA'] = homicidios_hecho['FECHA'].dt.strftime('%A')

#Creo un diccionario para modificar a español los nombres de los dás de la semana y lo aplico a la columna nueva
dic={  'Friday':'Viernes',
       'Saturday': 'Sabado',
       'Sunday': 'Domingo',  
       'Thursday': 'Jueves',
        'Monday': 'Lunes',
        'Wednesday': 'Miercoles',
       'Tuesday': 'Martes'}
homicidios_hecho['DIA SEMANA'] =homicidios_hecho['DIA SEMANA'].replace(dic)
```

Vamos a trabajar con los datos de Homicidio- Victimas.

Podemos observar que la fecha de fallecimiento  esta en formato object, debemos cambiarla a formato date.  tambien pudimo observar que hay datos faltantes en la columna en estudio, por tal motivo tomamos la desición de colocar la fecha de fallecimiento la misma fecha del sinietro, dando como conclusión que la victima falleció en el acto 

```python
homicidios_victima["FECHA_FALLECIMIENTO"] = pd.to_datetime(homicidios_victima["FECHA_FALLECIMIENTO"], errors='coerce')
homicidios_victima["FECHA_FALLECIMIENTO"].fillna(homicidios_victima["FECHA"], inplace=True)
```

```python
homicidios_victima.info()
```

Vamos a trabajar con la  EDAD ya que tiene formato ipo object. esta debe ser de tipo numerico.

```python
#Cambio la columna edad a numerico y luego a tipo entero
homicidios_victima["EDAD"] = pd.to_numeric(homicidios_victima["EDAD"], errors="coerce")
homicidios_victima["EDAD"] = homicidios_victima["EDAD"].astype("Int64")
```

Dado que no son tanto valores faltantes  en la edad de las personas. Por tal motivo  vamos a reemplazar por la media a estos valores.

```python
median_value = homicidios_victima["EDAD"].median()

# Replace missing values in "EDAD" column with the median value
homicidios_victima["EDAD"].fillna(median_value, inplace=True)
```

```python
homicidios_victima ["EDAD"].unique()
```

Vamos a crear una columna de rangos para estudiar su comportamiento por la edad segmentada, es decir,  menos de 20 años, entre 21 y 40 años, entre 41 y 60 años, entre 61 y 80 años y mas de 80 años. con esto podremos hacer analisis  mas exhaustivos.

```python
# Definir los límites de los rangos etarios y las etiquetas
rango = [0, 20, 40, 60, 80, 100]
etiquetas = ['Menos de 20', ' Entre 21- 40', 'Entre 41-60', 'Entre 61-80', 'Más de 80']

# Utilizar pd.cut() para crear la columna de rango etario
homicidios_victima['RANGO_EDAD'] = pd.cut(homicidios_victima['EDAD'], bins=rango , labels=etiquetas, right=False)

```

```python
homicidios_victima.head(3)
```

Vamos a eliminar las columnas que se repiten en el dataset de homicidios_hecho que son: FECHA, AÑO, MES, DIA Y VICTIMA.
